```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
url1 = 'AminoglycosidesBoth.csv'
#names = ['Aminoglycosides','Aminopenicillins','Amoxicillin','Cephalosporins','Fluoroquinolones']

df = pd.read_csv(url1)
df.head()

```

```python
y = df.AminoglycosidesRes
x = df.AminoglycosidesUse
x = sm.add_constant(x)
x.head()
```

```python
est = sm.OLS(y,x)
est = est.fit()
est.summary()
```
